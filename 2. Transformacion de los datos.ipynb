```python
#Importamos las librerias necesarias para transformar y modificar los datos

import numpy as np
import pandas as pd
```

```python
#Importamos el archivo .csv con read_csv

df = pd.read_csv("datos_climaticos.csv")
```

```python
df
```

```python
#Creamos la variable eliminar para eliminar las columnas que no necesitamos con drop

eliminar = ["n_fog","altitud","n_gra","np_100","np_001","np_300","n_tor","n_nie","np_010","nw_55","w_racha","nv_0050","ts_20","ts_10","ts_50","nw_91","w_rec","w_med","e","n_cub","n_des","n_nub","inso","q_max","q_mar","q_med","p_sol","nv_1000","nv_0100","q_min","evap","glo","indicativo","indsinop","p_max","ts_min","ta_min","ti_max","ta_max"]

df.drop(eliminar,axis = 1, inplace=True)
```

```python
df
```

```python
#Vamos a coger las siguientes estaciones para tener un plano general de España con sus coordenadas [Latitud, Longitud]:

#Hay que pasar de coordenadas UTM a coordenadas decimales para la proyeccion en la visualizacion

#1º "BARCELONA: BARCELONA AEROPUERTO" ["411734N","020412E"] = ["41.292777","2.070000"]
#2º "SEVILLA: SEVILLA AEROPUERTO" ["372500N","055245W"] = ["37.41666","-5.8791666"]
#3º "VALENCIA:VALENCIA" ["392850N","002159W"] = ["39.48055","-0.366388"]
#4º "BIZKAIA: BILBAO AEROPUERTO" ["431753N","025423W"] = ["43.298055","-2.906388"] 
#5º "MADRID: MADRID, RETIRO" ["402443N","034041W"] = ["40.411944","-3.678055"]  
#6º "OURENSE: OURENSE" ["421931N","075135W"] = ["42.325277","-7.859722"]  
#7º "ALMERIA: ALMERIA AEROPUERTO" ["365047N","022125W"] = ["36.846388","-2.356944"]
#8º "SALAMANCA: SALAMANCA AEROPUERTO" ["405734N","052954W"] = ["40.95944","-5.498333"]
#9º "ILLES BALEARS: PALMA DE MALLORCA, AEROPUERTO" ["393339N","024412E"] = ["39.560833","2.736666"]
#10º "STA. CRUZ DE TENERIFE: HIERRO AEROPUERTO" ["274908N","175320W"] = ["27.81888","-17.88888"]

```

```python
# Vamos a cambiar el nombre de las columnas con .rename:

nuevas_columnas = {
    "nt_30":"Dias>30º",
    "p_mes":"Precipitacion",
    "n_llu":"Lluvia",
    "latitud":"Latitud",
    "provincia":"Provincia",
    "nombre":"Nombre",
    "longitud":"Longitud",
    "fecha":"Fecha",
    "tm_min":"Tmed_min",
    "tm_mes":"Tmed_year",
    "tm_max":"Tmed_max",
    "hr":"Hum relativa",
    "nt_00":"Dias<0º"}

df = df.rename(columns = nuevas_columnas)
```

```python
df
```

```python
#Vamos a ordenar el DataFrame por grupos para que quede lo mas ordenado posible:

orden_columnas = ["Provincia","Nombre","Fecha","Latitud","Longitud","Tmed_min","Tmed_year","Tmed_max","Dias<0º","Dias>30º","Hum relativa","Precipitacion","Lluvia"]

df = df[orden_columnas]
```

```python
df
```

```python
# Una vez reordenadas las columnas, vamos a filtrar por las siguientes estaciones meterologicas:

""" 1º "BARCELONA: BARCELONA AEROPUERTO" 
    2º "SEVILLA: SEVILLA AEROPUERTO" 
    3º "VALENCIA:VALENCIA AEROPUERTO"
    4º "BIZKAIA: BILBAO AEROPUERTO" 
    5º "MADRID: MADRID, RETIRO" 
    6º "OURENSE: OURENSE" 
    7º  "STA. CRUZ DE TENERIFE: HIERRO AEROPUERTO"
    8º "SALAMANCA: SALAMANCA AEROPUERTO"
    9º "ILLES BALEARS: PALMA DE MALLORCA, AEROPUERTO" 
    """
```

```python
#Para filtrar los nombres de cada estacion, creamos primero una variable con todas las estaciones y luego creamos un filtrado nombrando la columna y añadiendolo con el .isin 

nombres_filtrar = ["BARCELONA AEROPUERTO","SEVILLA AEROPUERTO","VALENCIA AEROPUERTO","BILBAO AEROPUERTO","MADRID, RETIRO","OURENSE","MADRID, RETIRO","OURENSE","SALAMANCA AEROPUERTO", "PALMA DE MALLORCA, AEROPUERTO", "HIERRO AEROPUERTO"]

df_filtrado = df[df["Nombre"].isin(nombres_filtrar)]

df = df_filtrado

df
```

```python
#Creamos la columna Year para separar los meses de los años en la columna de la fecha:

df["Year"] = df["Fecha"].apply(lambda x : x.split("-")[0])

df.head()
```

```python
#En este codigo, hacemos una eliminacion del mes 13 para no tener la suma total anual a traves de un filtro:

df["mes_13"] = df["Fecha"].apply(lambda x : True if x.split("-")[1] == "13" else False)

df = df[df["mes_13"] == False]

df.drop("mes_13", axis = 1, inplace = True)

df.head()
```

```python
df = df.groupby(["Provincia", "Nombre", "Latitud", "Longitud", "Year"], as_index = False).agg({"Tmed_min" : "mean", "Tmed_year" : "mean", "Tmed_max" : "mean", "Dias<0º" : "sum",
                                                                                          "Dias>30º" : "sum", "Hum relativa" : "mean", "Precipitacion" : "sum", "Lluvia" : "sum"})
```

```python
#Variable Lluvia: Tomar datos a partir de 1985, debido a la falta de datos de algunas estaciones de años anteriores.

df
```

```python
#Vamos a redondear con dos decimales las variables de Temperatura y Humedad relativa

df["Tmed_min"] = df["Tmed_min"].round(2)
df["Tmed_year"] = df["Tmed_year"].round(2)
df["Tmed_max"] = df["Tmed_max"].round(2)
df["Hum relativa"] = df["Hum relativa"].round(2)
```

```python
df
```

```python
df = df.replace(np.nan, None)
```

```python
df
```

```python
# Creamos nuestro archivo final con los datos transformados y lo pasamos a un .csv

df_csv = 'df.csv'

df.to_csv(df_csv, index=False)
```
