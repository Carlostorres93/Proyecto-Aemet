```python
# En este codigo, vamos a explicar como extraer los datos de la API de Airtable:

def extraer_airtable():

    #Aqui ponemos algunos detalles para hacer la  autenticación en Airtable
    API_KEY = 'pat6Mbqhjg0DDmAHT.3aba72e1d9691fe48dfaa3661e9cc047037606aa8d1bfa8b9f39194af5c8fe96'
    BASE_ID = 'appd8sa59m5s3Cisn'
    TABLE_ID = 'tblR6GQlWMKKebPBS'

    #La URL estará compuesta por la url principal, la BASE_ID y la TABLA ID
    url = f'https://api.airtable.com/v0/{BASE_ID}/{TABLE_ID}'

    #Creamos el encabezado (headers) para crear una autorizacion a traves de nuestra API_KEY
    headers = {'Authorization': f'Bearer {API_KEY}',
               'Content-Type': 'application/json'
    }
    #Con params, añadimos el parametro offset, para que no tengamos limitaciones a la hora de extraer datos
    params = {"offset" : None}

    df_download = pd.DataFrame() 

    #Usamos while para obtener extraer los datos de la API Airtable
    
    while params.get("offset") != None or df_download.shape[0] == 0:

        response = requests.get(url, headers = headers, params = params) #Hacemos requests para recibir la peticion de la API

        if response.status_code == 200: #Si la respuesta es 200, los datos son descargados correctamente
            print("Se ha realizado la descarga completa de los datos")

        params["offset"] = response.json().get("offset") #Añadimos el parametro offset para eliminar la limitacion de extraer datos

        print(params.get("offset"))

        df_download = pd.concat([df_download, pd.json_normalize(response.json()["records"])], ignore_index = True) #Hacemos concdenacion de los archivos descargados, normalizando los archivos json e ignorando los indices de nuestra descarga.

        # sleep(0.7) #Se hace sleep para no saturar la API
        
    return df_download #Devuelve los archivos que estamos descargando
```
